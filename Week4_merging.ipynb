```python
t3d_value_counts = pd.read_csv('./t3d_value_counts.csv')
t7d_value_counts = pd.read_csv('./t7d_value_counts.csv')
control_value_counts = pd.read_csv('./value_counts_control.csv')
```

```python
t3d_value_counts.head()
```

```python
t7d_value_counts.head()
```

```python
control_value_counts.head()
```

```python
t3d_value_counts = t3d_value_counts.rename(columns={'count': 'count_t3d'})
t7d_value_counts = t7d_value_counts.rename(columns={'count': 'count_t7d'})
control_value_counts = control_value_counts.rename(columns={'count': 'count_control'})
```

```python
combined_counts = pd.merge(t3d_value_counts, t7d_value_counts, on='gene_name', how='outer')
combined_counts = pd.merge(combined_counts, control_value_counts, on='gene_name', how='outer')
```

```python
combined_counts.head()
```

```python
combined_counts = combined_counts.fillna(0)
```

```python
combined_counts[['count_t3d', 'count_t7d', 'count_control']] = combined_counts[['count_t3d', 'count_t7d', 'count_control']].astype(int)
```

```python
combined_counts.head()
```

```python
combined_counts = combined_counts.sort_values(by='count_control', ascending=False)
```

```python
combined_counts.to_csv("./value_counts_combined.csv")
```
